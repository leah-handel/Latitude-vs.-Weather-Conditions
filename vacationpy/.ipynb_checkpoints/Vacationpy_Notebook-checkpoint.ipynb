```python
import pandas as pd
import numpy as np
import requests
import json
from scipy.stats import linregress
from citipy import citipy
import random
from config import gkey
import time
import matplotlib.pyplot as plt
import gmaps
```

```python
gmaps.configure(api_key=gkey)
```

```python
weather_url = "../weatherpy/results/weather_data.csv"
```

```python
weather = pd.read_csv(weather_url)
```

```python
weather.head()
```

```python
weather.dtypes
```

```python
locations = weather[["latitude", "longitude"]]
```

```python
humidity = weather["humidity"].astype(float)
```

```python
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)

fig.add_layer(heat_layer)

fig
```

```python
vacation_spots = weather.loc[(weather["temperature"]>77)&
            (weather["temperature"]<83)&
            (weather["cloudiness"]<20)&
            (weather["wind"]<10)&
            (weather["humidity"]>40)&
            (weather["humidity"]<80),:]
```

```python
vacation_spots
```

```python
vacation_spots["hotel name"]=""

vacation_spots
```

```python
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

location = "8.8978,35.1478"

params = {
    "location": location,
    "radius": "10000",
    "keyword": "hotel",
    "key": gkey,
}

response = requests.get(url, params=params).json()
```

```python
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
#hotel name
response["results"][0]["name"]
```

```python
url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

for index, row in vacation_spots.iterrows():
    
    location = f"{row['latitude']}, {row['longitude']}"
    
    params = {
    "location": location,
    "radius": "10000",
    "keyword": "hotel",
    "key": gkey}
    
    response = requests.get(url, params=params).json()
    
    vacation_spots.loc[index, "hotel name"] = response["results"][0]["name"]
    
```
